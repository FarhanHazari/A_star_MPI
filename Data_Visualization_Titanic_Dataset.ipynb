# ** Data Visualization of Titanic Dataset**

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data
data=sns.load_dataset('titanic')
data
```

```python
#simple line plot
plt.plot(data['age'],data['fare'])
plt.show
```

```python
#Customizing graph
plt.plot(data['age'],data['fare'],color="red")
plt.xlabel("Age of Passengers")
plt.ylabel("Fare")
plt.title("Age-Fare")
plt.show
```
